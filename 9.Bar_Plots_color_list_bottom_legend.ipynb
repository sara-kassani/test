```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams.update({'font.size': 16})
```

```python

colors_list = ['#ff1100','#0800ff','#00bfff', '#fa9600', '#ff00fb']

Xception  = [79.59, 82.35, 86.32]
ResNet50 = [74.64, 56.52, 85.71]
MobileNet = [79.01, 76.47, 84.62]
InceptionV3 = [78.72, 63.64, 85.37]
Modified_Xception = [83.09, 88.24, 87.00]

index = ['Accuracy','Sensitivity', 'Specificity']

df = pd.DataFrame({'Xception': Xception, 'ResNet50': ResNet50, 'MobileNet': MobileNet,
                   'InceptionV3': InceptionV3, 'Modified Xception': Modified_Xception}, index=index)



ax = df.plot.bar(color=colors_list, rot=0, figsize=(10, 6))
plt.ylim(50,90)

# plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.)

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08), ncol=5, prop={'size': 14})
```
